1. Buatlah dataframe yang disimpan dalam variabel 'df_england', yang merupakan data pemain dengan kewarganegaraan England

5. Buatlah kolom baru yang bernama "players category" yang berisi kategori players sebagai berikut :
- "Legend" : Jika overall bernilai lebih dari 90
- "High Class" : Jika overall bernilai dari 80 sampai 90 
- "Great" : Jika overall bernilai dari 70 sampai 80
- "Average" : Jika overall bernilai dari 60 sampai 70
- "Low Class" : Jika overall bernilai dari 50 sampai 60
- "Bottom Class" : Jika overall bernilai dibawah 50

```python
def players_category(x):
    if x['overall'] > 90:
        segment = 'Legend'
    elif x['overall'] > 80 and x['overall'] <= 90:
        segment = 'High Class'
    elif x['overall'] > 70 and x['overall'] <= 80:
        segment = 'Great'
    elif x['overall'] >= 60 and x['overall'] <= 70:
        segment = 'Average'
    elif x['overall'] >= 50 and x['overall'] <= 60:
        segment = 'Low Class'
    elif x['overall'] < 50:
        segment = 'Bottom Class'
    return segment

df_england['players_category'] = df_england.apply(lambda x: players_category(x), axis=1)

df_england.head(3)
```

6. Ada berapakah pemain England yang masuk kategori "Bottom Class" ?

```python
df_england[(df_england['nationality']=='England') & (df_england['players_category']=='Bottom Class')].count() 
```
